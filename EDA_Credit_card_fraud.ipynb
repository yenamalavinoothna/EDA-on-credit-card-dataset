```python
# Importing
import sklearn

# IQR
Q1 = np.percentile(data['V15'], 25,interpolation = 'midpoint')

Q3 = np.percentile(data['V15'], 75,interpolation = 'midpoint')
IQR = Q3 - Q1

print("Old Shape: ", data.shape)

# Upper bound
upper = np.where(data['V15'] >= (Q3+1.5*IQR))

# Lower bound
lower = np.where(data['V15'] <= (Q1-1.5*IQR))

# Removing the Outliers
data.drop(upper[0], inplace = True)
data.drop(lower[0], inplace = True)

print("New Shape: ", data.shape)

sns.boxplot(x='V15', data=data)
```
